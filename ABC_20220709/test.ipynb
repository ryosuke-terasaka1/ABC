```python
def decimal_to_biquinary_notation(num: int):
    p = num
    q = num 
    result = ''
    while p >= 1:
        q = p % 9
        p //= 9
        if q == 8:
            result = '5' + result
        else:
            result = str(q) + result
        print(p, q)
    return int(result)
print(decimal_to_biquinary_notation(107))
```
