```python
n, m = map(int, input().split())

x = list(map(int, input().split()))

dp = [0] * (n)
dp[0] = x[0]
bonus_dic = {}
bonus_sum = 0

for _ in range(m):
    c, y = map(int, input().split())
    bonus_dic[c] = y
    bonus_sum += y

for i in range(1, n):
    bonus_sum = 0
    for k, v in bonus_dic.items():
        if i == k:
            dp[i] = max(dp[i-1]+x[i], x[i]+v)
        elif i > k > 1:
            dp[i] = max(dp[i-k-1]+sum(x[i-k+1:i+1])+v, dp[i], dp[i-1]+x[i]+bonus_sum)
        else:
            dp[i] = dp[i-1]+x[i]
print(dp)
print(x)
print(max(dp[n-1], bonus_sum+sum(x)))
```

```python
def decimal_to_biquinary_notation(num: int):
    p = num
    q = num 
    result = ''
    while p >= 1:
        q = p % 9
        p //= 9
        if q == 8:
            result = '5' + result
        else:
            result = str(q) + result
        print(p, q)
    return int(result)
print(decimal_to_biquinary_notation(107))
```

```python
from typing import List
import itertools

part_list = ['ボーカル', 'メロディ', 'ドラム']
num_list = [0, 1]
key_list = []
value_list = {}

for i, j, k in itertools.product(num_list, num_list, num_list):
    value_list[(i, j, k)] = ""
    key_list.append((i, j, k))

for a,b,c,d,e,f,g,h in itertools.product(part_list, part_list, part_list, part_list, part_list, part_list, part_list, part_list):
    rule_list = [a,b,c,d,e,f,g,h]
    for i in range(len(rule_list)):
        value_list[key_list[i]] = rule_list[i]
print(value_list[(0, 0, 0)])
# print(key_list)
```
